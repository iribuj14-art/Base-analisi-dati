```python
"""
analisi geo loc.

considerazione dei primi 10 paesi

"""

import os
import re
import csv
import time
import logging
from collections import Counter, defaultdict

import pandas as pd
import matplotlib.pyplot as plt

# import pip
from rapidfuzz import process, fuzz

# offline dataset
import geonamescache


from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
import requests_cache

# Config
INPUT_CSV = "datasets/cybersecurity_attacks.csv"                
LOCATION_COL = "Geo-location Data"    
CACHE_DB = "geocode_cache"            
PLOT_FILE = "top10_countries.png"
TOP_N = 10


logging.basicConfig(level=logging.INFO, format="%(levelname)s: %(message)s")


def normalize(text):
    if not isinstance(text, str):
        return ""
    text = text.strip()
    text = re.sub(r"\s+", " ", text)
    text = re.sub(r"[^\w\s,-]", "", text)
    return text

# Parsing
def parse_location_field(s):
    """
    Returns (city, region) where either may be None.
    Example inputs:
      'Jamshedpur, Sikkim' -> ('Jamshedpur', 'Sikkim')
      'Paris' -> ('Paris', None)
      'San Jose, CA' -> ('San Jose', 'CA')
    """
    s = normalize(s)
    if not s:
        return (None, None)
    parts = [p.strip() for p in s.split(",") if p.strip()]
    if len(parts) == 1:
        return (parts[0], None)
    else:
        # city is first, rest join as region
        return (parts[0], ", ".join(parts[1:]))

# offline mapping
def build_offline_city_index():
    """
    Build maps:
      - city_name_lower -> list of (cityid, cityinfo)
      cityinfo contains name, countrycode, population, admin1 code etc.
    """
    gc = geonamescache.GeonamesCache()
    cities = gc.get_cities()  
    city_index = defaultdict(list)
    for geoid, info in cities.items():
        name = info.get("name", "")
        if not name:
            continue
        city_index[name.lower()].append(info)
    return city_index

def offline_lookup_country(city_name, region, city_index):
    if not city_name:
        return None, "no_city"
    key = city_name.lower()
    candidates = city_index.get(key)
    if not candidates:
        choices = list(city_index.keys())
        best = process.extractOne(key, choices, scorer=fuzz.WRatio, score_cutoff=80)
        if best:
            matched_key = best[0]
            candidates = city_index.get(matched_key)
            reason = f"fuzzy_matched:{matched_key}"
        else:
            return None, "no_match_offline"
    else:
        reason = "exact_name"

    if len(candidates) == 1:
        info = candidates[0]
        return info.get("countrycode"), reason

    if region:
        rlow = region.lower()
        for info in candidates:
            admin1 = info.get("admin1")
            if admin1 and rlow in admin1.lower():
                return info.get("countrycode"), f"{reason}_matched_region"
        gc = geonamescache.GeonamesCache()
        countries = gc.get_countries()
        for info in candidates:
            cc = info.get("countrycode")
            ctry = countries.get(cc, {}).get("name", "")
            if ctry and rlow in ctry.lower():
                return cc, f"{reason}_matched_countryname_region"
    pop_sorted = sorted(candidates, key=lambda x: int(x.get("population") or 0), reverse=True)
    return pop_sorted[0].get("countrycode"), f"{reason}_popfallback"

class OnlineGeocoder:
    def __init__(self, user_agent="geo_script"):
        requests_cache.install_cache(CACHE_DB)
        self.geolocator = Nominatim(user_agent=user_agent, timeout=10)
        self.geocode = RateLimiter(self.geolocator.geocode, min_delay_seconds=1, max_retries=2, error_wait_seconds=2)

    def geocode_to_country(self, q):
        if not q:
            return None, "no_query"
        try:
            loc = self.geocode(q, addressdetails=True)
        except Exception as e:
            return None, f"geopy_error:{e}"
        if not loc:
            return None, "no_result"
        raw = loc.raw
        addr = raw.get("address", {})
        cc = addr.get("country_code")
        country = addr.get("country")
        if cc:
            return cc.upper(), f"online_country:{country}"
        return None, "online_no_cc"

#pincipale
def geolocate_dataframe(df, offline_first=True, allow_online_fallback=True):
    city_index = build_offline_city_index()
    geocoder = None
    results = []
    for i, val in enumerate(df[LOCATION_COL].fillna("")):
        city, region = parse_location_field(val)
        country = None
        reason = None
        
        if offline_first:
            country, reason = offline_lookup_country(city, region, city_index)
        if (not country) and allow_online_fallback:
            if geocoder is None:
                geocoder = OnlineGeocoder()
            query_parts = []
            if city:
                query_parts.append(city)
            if region:
                query_parts.append(region)
            query_parts.append(val)
            query = ", ".join([p for p in query_parts if p])
            country, reason = geocoder.geocode_to_country(query)

        results.append({
            "original": val,
            "city": city,
            "region": region,
            "country_code": country,
            "reason": reason
        })
        if (i + 1) % 500 == 0:
            logging.info(f"Processed {i+1} rows")

    return pd.DataFrame(results)

def country_code_to_name(code):
    if not code:
        return None
    gc = geonamescache.GeonamesCache()
    countries = gc.get_countries()
    c = countries.get(code)
    if c:
        return c.get("name")
    return code

def plot_top_countries(country_names, top_n=10, file_out=PLOT_FILE):
    cnt = Counter(country_names)
    if None in cnt:
        del cnt[None]
    most = cnt.most_common(top_n)
    if not most:
        print("No countries found to plot.")
        return
    countries, counts = zip(*most)

    plt.figure(figsize=(10, 6))
    bars = plt.bar(range(len(countries)), counts)
    plt.xticks(range(len(countries)), countries, rotation=45, ha="right")
    plt.ylabel("Count")
    plt.title(f"Top {len(countries)} countries in dataset")
    plt.tight_layout()
    plt.savefig(file_out, dpi=150)
    plt.show()
    print(f"Saved plot to {file_out}")

def main():
    if not os.path.exists(INPUT_CSV):
        raise FileNotFoundError(f"Input CSV not found: {INPUT_CSV}")

    df = pd.read_csv(INPUT_CSV, dtype=str)
    if LOCATION_COL not in df.columns:
        raise KeyError(f"Column '{LOCATION_COL}' not found in CSV. Columns: {df.columns.tolist()}")

    logging.info(f"Loaded {len(df)} rows from {INPUT_CSV}")
    results_df = geolocate_dataframe(df, offline_first=True, allow_online_fallback=True)
    results_df["country_name"] = results_df["country_code"].apply(country_code_to_name)

    results_df.to_csv("geolocated_results.csv", index=False)
    logging.info("Wrote per-row geolocation results to geolocated_results.csv")

    plot_top_countries(results_df["country_name"].tolist(), top_n=TOP_N, file_out=PLOT_FILE)

if __name__ == "__main__":
    main()
```

```python
import os
import json
import math
from pathlib import Path
from IPython.display import IFrame, display, HTML

import pandas as pd
import folium
from folium.plugins import MarkerCluster, HeatMap
import geoip2.database
from geoip2.errors import AddressNotFoundError
import ipaddress
from tqdm.notebook import tqdm

GEOIP_DB_PATH = "GeoLite2-City.mmdb"
CACHE_PATH = "ipgeo_cache.json"
ROUND_PRECISION = 4     
HEATMAP = True          
OUT_PREFIX = "ipmaps"  

def is_public_ip(ip_str):
    try:
        ip = ipaddress.ip_address(ip_str)
    except Exception:
        return False
    if ip.is_private or ip.is_loopback or ip.is_link_local or ip.is_multicast or ip.is_unspecified or ip.is_reserved:
        return False
    return True

def load_cache(path):
    if os.path.exists(path):
        try:
            with open(path, "r", encoding="utf-8") as f:
                return json.load(f)
        except Exception:
            return {}
    return {}

def save_cache(path, cache):
    tmp = path + ".tmp"
    with open(tmp, "w", encoding="utf-8") as f:
        json.dump(cache, f, ensure_ascii=False)
    os.replace(tmp, path)

def lookup_ips_maxmind(unique_ips, db_path, cache, show_progress=True):
    ips_to_lookup = [ip for ip in unique_ips if ip not in cache and is_public_ip(ip)]
    if not ips_to_lookup:
        return cache
    reader = geoip2.database.Reader(db_path)
    iterator = tqdm(ips_to_lookup, desc="Geolocalizzazione IP", unit="ip") if show_progress else ips_to_lookup
    for ip in iterator:
        try:
            rec = reader.city(ip)
            lat = rec.location.latitude
            lon = rec.location.longitude
            city = rec.city.name if rec.city and rec.city.name else None
            country = rec.country.iso_code if rec.country and rec.country.iso_code else None
            subdivision = None
            if rec.subdivisions and rec.subdivisions.most_specific:
                subdivision = rec.subdivisions.most_specific.name
            if lat is None or lon is None:
                cache[ip] = None
            else:
                cache[ip] = {"lat": float(lat), "lon": float(lon), "city": city, "country": country, "subdivision": subdivision}
        except AddressNotFoundError:
            cache[ip] = None
        except Exception as e:
            cache[ip] = None
    reader.close()
    return cache

def build_aggregated_dataframe(ip_list, cache, round_precision=ROUND_PRECISION):
    rows = []
    for ip in ip_list:
        if pd.isna(ip):
            continue
        ip = str(ip).strip()
        loc = cache.get(ip)
        if not loc:
            continue
        lat = loc["lat"]; lon = loc["lon"]
        rlat = round(lat, round_precision); rlon = round(lon, round_precision)
        rows.append({"ip": ip, "lat": lat, "lon": lon, "rlat": rlat, "rlon": rlon, "city": loc.get("city"), "country": loc.get("country")})
    if not rows:
        return pd.DataFrame(columns=["rlat","rlon","count","ips","cities","countries","lat_mean","lon_mean"])
    d = pd.DataFrame(rows)
    agg = d.groupby(["rlat","rlon"]).agg(
        count=("ip","count"),
        ips=("ip", lambda s: ", ".join(list(s.unique()[:10]))),
        cities=("city", lambda s: ", ".join([str(x) for x in s.dropna().unique()[:3]])),
        countries=("country", lambda s: ", ".join([str(x) for x in s.dropna().unique()[:3]])),
        lat_mean=("lat", "mean"),
        lon_mean=("lon", "mean")
    ).reset_index()
    return agg

def create_and_show_map(agg_df, out_html, title="IP map", heatmap=HEATMAP, start_zoom=2, width=900, height=500):
    if agg_df.empty:
        m = folium.Map(location=[20,0], zoom_start=1)
        m.save(out_html)
        print(f"[WARN] No public geolocated IPs to plot. Saved empty map to {out_html}")
        display(IFrame(out_html, width=width, height=height))
        return
    center_lat = agg_df["lat_mean"].mean()
    center_lon = agg_df["lon_mean"].mean()
    m = folium.Map(location=[center_lat, center_lon], zoom_start=start_zoom)
    folium.TileLayer("OpenStreetMap").add_to(m)
    mc = MarkerCluster(name="IP clusters")
    m.add_child(mc)
    for _, row in agg_df.iterrows():
        lat = row["rlat"]; lon = row["rlon"]
        popup_html = f"<b>Count:</b> {row['count']}<br><b>IPs:</b> {row['ips']}<br><b>Cities:</b> {row['cities']}<br><b>Countries:</b> {row['countries']}"
        popup = folium.Popup(popup_html, max_width=400)
        radius = 4 + math.log(row["count"] + 1) * 3
        folium.CircleMarker(location=[lat, lon], radius=radius, weight=1, fill=True, fill_opacity=0.6, popup=popup).add_to(mc)
    if heatmap:
        heat_data = [[r["rlat"], r["rlon"], r["count"]] for _, r in agg_df.iterrows()]
        HeatMap(heat_data, name="Heatmap", radius=15).add_to(m)
    m.save(out_html)
    print(f"Saved map to {out_html}")
    display(IFrame(out_html, width=width, height=height))
```

```python
src_ips = df[SRC_COL].dropna().tolist()
dst_ips = df[DST_COL].dropna().tolist()
unique_ips = set([ip for ip in src_ips + dst_ips if ip and not pd.isna(ip)])

print(f"Unique IPs to consider: {len(unique_ips)}")

cache = load_cache(CACHE_PATH)
cache = lookup_ips_maxmind(unique_ips, GEOIP_DB_PATH, cache, show_progress=True)
save_cache(CACHE_PATH, cache)
print(f"Cache saved to {CACHE_PATH}")

agg_src = build_aggregated_dataframe(src_ips, cache, round_precision=ROUND_PRECISION)
agg_dst = build_aggregated_dataframe(dst_ips, cache, round_precision=ROUND_PRECISION)

out_src = f"{OUT_PREFIX}_source.html"
out_dst = f"{OUT_PREFIX}_destination.html"

create_and_show_map(agg_src, out_src, title="Source IPs", heatmap=HEATMAP)
create_and_show_map(agg_dst, out_dst, title="Destination IPs", heatmap=HEATMAP)
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display, Image, HTML

CSV_PATH = globals().get("CSV_PATH", "data.csv")
CACHE_PATH = globals().get("CACHE_PATH", "ipgeo_cache.json")
SRC_COL = globals().get("SRC_COL", "Source IP Address")
DST_COL = globals().get("DST_COL", "Destination IP Address")
TOP_N = 10

out_src_png = "top10_countries_source.png"
out_dst_png = "top10_countries_destination.png"
out_both_png = "top10_countries_both.png"

if "df" not in globals():
    if os.path.exists(CSV_PATH):
        df = pd.read_csv(CSV_PATH, dtype=str, low_memory=False)
        df[SRC_COL] = df[SRC_COL].astype(str).str.strip()
        df[DST_COL] = df[DST_COL].astype(str).str.strip()
    else:
        raise SystemExit("Dataframe 'df' not found in memory and CSV_PATH file not found. Run earlier cells that load your CSV.")

if "cache" not in globals():
    if os.path.exists(CACHE_PATH):
        import json
        with open(CACHE_PATH, "r", encoding="utf-8") as f:
            cache = json.load(f)
    else:
        raise SystemExit("IP geolocation cache not found. Run the geolocation step to populate ipgeo_cache.json or set CACHE_PATH.")

def ip_to_country(ip):
    if not isinstance(ip, str) or not ip:
        return "Unknown"
    loc = cache.get(ip)
    if not loc:
        return "Unknown"
    return loc.get("country") or "Unknown"

src_countries = df[SRC_COL].dropna().astype(str).map(ip_to_country)
dst_countries = df[DST_COL].dropna().astype(str).map(ip_to_country)

def top_n_table(series, n=TOP_N):
    counts = series.value_counts(dropna=False)
    total = counts.sum()
    top = counts.head(n).rename_axis("country").reset_index(name="count")
    top["percent"] = (top["count"] / total * 100).round(2)
    top["country"] = top["country"].fillna("Unknown").astype(str)
    return top, total

top_src, total_src = top_n_table(src_countries, TOP_N)
top_dst, total_dst = top_n_table(dst_countries, TOP_N)

top_src.to_csv(f"top_{TOP_N}_countries_source.csv", index=False)
top_dst.to_csv(f"top_{TOP_N}_countries_destination.csv", index=False)

def plot_vertical_bar(df_top, title, out_png, rotate_xticks=45, figsize=None):
    if df_top.empty:
        print(f"No data to plot for {title}.")
        plt.figure(figsize=(6,4))
        plt.text(0.5, 0.5, "No data", ha="center", va="center")
        plt.axis("off")
        plt.savefig(out_png, bbox_inches="tight", dpi=150)
        plt.close()
        return
    labels = df_top["country"].tolist()
    values = df_top["count"].tolist()
    n = len(labels)
    width = max(6, n * 0.8)
    height = 6 if figsize is None else figsize[1]
    plt.figure(figsize=(width, height))
    plt.bar(labels, values)
    plt.xlabel("Paese")
    plt.ylabel("Quantità IP")
    plt.title(title)
    plt.xticks(rotation=rotate_xticks, ha="right")
    plt.grid(axis="y", linestyle=":", alpha=0.6)
    plt.tight_layout()
    plt.savefig(out_png, bbox_inches="tight", dpi=150)
   # plt.show()
    plt.close()

plot_vertical_bar(top_src, f"Primi {TOP_N} Paesi — IP di origine (n={total_src})", out_src_png)

plot_vertical_bar(top_dst, f"Primi {TOP_N} Paesi — IP di arrivo (n={total_dst})", out_dst_png)

def plot_combined(df1, df2, title1, title2, out_png, rotate_xticks=45):
    n1 = len(df1); n2 = len(df2)
    figsize = (max(10, max(n1, n2) * 0.7), 6)
    fig, axes = plt.subplots(1, 2, figsize=figsize)
    axes[0].bar(df1["country"], df1["count"], color="#430863")
    axes[0].set_title(title1)
    axes[0].set_xlabel("Paese", fontsize=18)
    axes[0].set_ylabel("Quantità IP", fontsize=18)
    axes[0].tick_params(axis='x', rotation=rotate_xticks)
    axes[0].grid(axis="y", linestyle=":", alpha=0.6)
    axes[1].bar(df2["country"], df2["count"], color="#790399")
    axes[1].set_title(title2)
    axes[1].set_xlabel("Paese", fontsize=14)
    axes[1].set_ylabel("Quantità IP", fontsize=14)
    axes[1].tick_params(axis='x', rotation=rotate_xticks)
    axes[1].grid(axis="y", linestyle=":", alpha=0.6)
    plt.subplots_adjust(wspace=0.6)
    plt.savefig(out_png, bbox_inches="tight", dpi=150)
    plt.show()
    plt.close()

plot_combined(top_src, top_dst, f"Origine (top {TOP_N})", f"Arrivo (top {TOP_N})", out_both_png)


```

```python
import os
import pandas as pd
from IPython.display import display, HTML

CSV_PATH = globals().get("CSV_PATH", "data.csv")
DF_VAR = globals().get("df", None)  
TIMESTAMP_COL = "Timestamp"         

if DF_VAR is None:
    if os.path.exists(CSV_PATH):
        df = pd.read_csv(CSV_PATH, dtype=str, low_memory=False)
        print(f"Loaded {CSV_PATH} ({len(df)} rows).")
    else:
        raise SystemExit("Dataframe not found in memory and CSV_PATH file not found. Run earlier cells that load your CSV.")
else:
    df = DF_VAR
    print(f"Using in-memory dataframe (rows={len(df)}).")

if TIMESTAMP_COL not in df.columns:
    raise SystemExit(f"Column '{TIMESTAMP_COL}' not found. Available columns: {list(df.columns)}")

fmt = "%Y-%m-%d %H:%M:%S"
df["_ts_parsed"] = pd.to_datetime(df[TIMESTAMP_COL], format=fmt, errors="coerce")

if df["_ts_parsed"].isna().any():
    n_bad = df["_ts_parsed"].isna().sum()
    print(f"{n_bad} rows failed strict parsing; attempting flexible parse for those rows...")
    mask_bad = df["_ts_parsed"].isna()
    df.loc[mask_bad, "_ts_parsed"] = pd.to_datetime(df.loc[mask_bad, TIMESTAMP_COL], errors="coerce", utc=False)

n_invalid = df["_ts_parsed"].isna().sum()
if n_invalid:
    print(f"Warning: {n_invalid} rows have invalid timestamps and will be ignored in time-series plots.")

display(HTML(f"<b>Parsed timestamps:</b> earliest = {df['_ts_parsed'].min()}, latest = {df['_ts_parsed'].max()}"))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2_contingency, circmean, circstd

df = pd.read_csv("datasets/cybersecurity_attacks.csv") 

df['Timestamp'] = pd.to_datetime(df['Timestamp'])

df['hour'] = df['Timestamp'].dt.hour
df['day_of_week'] = df['Timestamp'].dt.day_name()
df['date'] = df['Timestamp'].dt.date

attacks_per_hour = df.groupby('hour').size()
days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
attacks_per_day = df.groupby('day_of_week').size().reindex(days_order)
attacks_per_date = df.groupby('date').size()

plt.figure(figsize=(10,5))
attacks_per_hour.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Attacco per ora')
plt.xlabel('Orario')
plt.ylabel('Numero attacchi')
plt.xticks(rotation=0)
plt.show()

plt.figure(figsize=(10,5))
attacks_per_day.plot(kind='bar', color='salmon', edgecolor='black')
plt.title('Attacco per giorno della settimana')
plt.xlabel('Giorno della settimana')
plt.ylabel('Numero attacchi')
plt.xticks(rotation=0)
plt.show()

heatmap_data = df.groupby(['day_of_week', 'hour']).size().unstack().reindex(days_order)
plt.figure(figsize=(12,6))
sns.heatmap(heatmap_data, cmap="YlGnBu", annot=True, fmt='d')
plt.title('Heatmap degli attacchi (giorno vs ora)')
plt.xlabel('Ora')
plt.ylabel('Giorno')
plt.show()

contingency_table = attacks_per_day.values.reshape(1, -1)
chi2, p, dof, expected = chi2_contingency(contingency_table)
print(f"Chi-square test p-value for attacks across days: {p:.4f}")
if p < 0.05:
    print("Significant difference: attacks are not evenly distributed across days.")
else:
    print("No significant difference: attacks are evenly distributed across days.")

hours_rad = df['hour'] * (2 * np.pi / 24)  
mean_hour = circmean(hours_rad, high=2*np.pi, low=0) * 24 / (2*np.pi)
std_hour = circstd(hours_rad, high=2*np.pi, low=0) * 24 / (2*np.pi)
print(f"Circular mean hour: {mean_hour:.2f}")
print(f"Circular std hour: {std_hour:.2f}")

plt.figure(figsize=(8,6))
plt.hist(hours_rad, bins=24, density=True, color='purple', edgecolor='black')
plt.title("Istogramma degli orari")
plt.xlabel("Ora (radianti)")
plt.ylabel("Densità")
plt.show()

import numpy as np

def rayleigh_test(angles_rad):
    n = len(angles_rad)
    R = np.sqrt(np.sum(np.cos(angles_rad))**2 + np.sum(np.sin(angles_rad))**2) / n
    p_value = np.exp(-n * R**2)
    return R, p_value

R, p_value = rayleigh_test(hours_rad)
print(f"Rayleigh test statistic R: {R:.4f}")
print(f"Rayleigh test p-value: {p_value:.4f}")

if p_value < 0.05:
    print("Significant clustering: attacks are more likely at certain hours.")
else:
    print("No significant clustering: attacks are uniform across hours.")
```

```python
hours_rad = df.groupby('hour').size().index.values * 2 * np.pi / 24
counts = df.groupby('hour').size().values

sin_hours = np.sin(hours_rad)
cos_hours = np.cos(hours_rad)

from scipy.stats import pearsonr
r_sin, p_sin = pearsonr(sin_hours, counts)
r_cos, p_cos = pearsonr(cos_hours, counts)

r = np.sqrt(r_sin**2 + r_cos**2)
print(f"Circular-linear correlation r: {r:.4f}")
```
